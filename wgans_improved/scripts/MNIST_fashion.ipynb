```python
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
%load_ext autoreload
%autoreload 2
from utils import *
from MLP_WGAN import MLP_WGAN
from DCGAN import DC_WGAN
```

## Load data

```python
mnist = input_data.read_data_sets("FashionMNIST_data/", one_hot=True, reshape=[])
train_set = mnist.train.images

print(train_set.shape)
```

## Training

### Defining Parameters

```python
mb_size = 64
z_dim = 100
n_fig= 16
n_disc = 5
train_epoch = 5000
penalty = 10
x_dim = train_set.shape[1]
color = train_set.shape[3]
```

### DCGAN architecture
Training and saving the model

```python
out_dir = 'out_DCGAN_fashion'
mnist_train_dcgan = DC_WGAN(train_set, mb_size, train_epoch, x_dim, z_dim, n_fig, out_dir, color, penalty, n_disc)
history_dcgan=mnist_train_dcgan.training()
```

### Plotting Output
Plotting last 10 images

```python
fileNames=[]
for root, dirs, files in os.walk(out_dir):
    for filename in files:
        if os.path.splitext(filename)[1].lower() =='.png':
                print(root+'/'+filename)
                fileNames.append(root+'/'+filename)
```

```python
last_n=5
for i in range(-1, -last_n-1, -1):
    print(fileNames[i])
    img=mpimg.imread(fileNames[i])
    fig = plt.imshow(img)
    fig.axes.axis('off')
    plt.show()
    plt.close()
```

### Restoring Saved Models

```python
generateSamples(out_dir)
```

### Generator & Discriminator Loss Curves

```python
G = history_mlp[1]
D = history_mlp[0]

# generator
plt.figure(figsize=(9, 6))
plt.plot(range(0,5000,100), G)
plt.title("Generator Training Loss")
plt.ylabel('Loss')
plt.xlabel('iteration')
plt.show()

# discriminator
plt.figure(figsize=(9, 6))
plt.plot(range(0,5000,100), np.multiply(D,-1), color='orange')
plt.title("Wasserstein Distance")
plt.ylabel('Wasserstein Distance')
plt.xlabel('iteration')
plt.show()
```
